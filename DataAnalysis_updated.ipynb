###### Data cleaning for 2018

##### Data cleaning for 2019

```python
data2019.head()
```

```python
# Remove special characters in the Amount($) column
data2019['Amount($)'] = data2019['Amount($)'].str.replace(r'[$₹]', '', regex=True)
data2019['Amount($)'] = data2019['Amount($)'].str.replace(',', '', regex=True)
```

```python
# Convert Amount($) column to float datatype and convert the Founded column to int datatype
data2019['Amount($)'] = pd.to_numeric(data2019['Amount($)'], errors='coerce')
data2019['Founded'] = pd.to_datetime(data2019['Founded'], format='%Y').dt.year
```

```python
# Find the sum of null values for each variable(column)
null_counts = data2019.isnull().sum()
print(null_counts)
```

```python
# Fill missing values in the Amount($) column with the mean of that column (data imputation)
data2019['Amount($)'].fillna(data2019['Amount($)'].mean(), inplace=True)
```

```python
# Filter/Sort out the null values in the sector column
filtered_df = data2019[data2019['Sector'].isnull()]
filtered_df
```

```python
# Drop row 41 as 55% of observations or values in that row are missing
data2019.drop(index=41, inplace=True)
```

```python
# Find out the unique values in the Sector column
unique_values_sector = data2019['Sector'].unique()
print(unique_values_sector)
```

```python
# Replace non values with similar observations in the rest of the data set (data imputation)
data2019.loc[41, 'Sector'] = 'Social Media'
data2019.loc[49, 'Sector'] = 'B2B'
data2019.loc[52, 'Sector'] = 'Social Media'
data2019.loc[56, 'Sector'] = 'Fintech'
data2019.loc[57, 'Sector'] = 'E-marketplace'
```

```python
# Drop Stage column as more than 50% of observations in the Stage column is missing.
data2019.drop('Stage', axis=1, inplace=True)
```
